```python
import requests
import requests_cache
from bs4 import BeautifulSoup
import time
import json
import re
import pandas as pd # Dùng để đọc và hiển thị file csv/tsv
from datetime import datetime, timedelta # Dùng để xử lý dữ liệu thời gian
# YOUR CODE HERE (OPTION) 
# Nếu cần các thư viện khác thì bạn có thể import ở đây
# from requests_html import HTMLSession

import urllib.robotparser # Kiểm tra file robot.txt có được phép crawl không
```

```python
url = f'https://www.worldometers.info/coronavirus'
now_r = requests.get(url)

now_soup = BeautifulSoup(now_r.text, 'lxml')
now_table = now_soup.find('table', attrs={'id':'main_table_countries_today'})
now_table_header = now_table.find('thead')
now_table_body = now_table.find('tbody')

now_attr_list = [str(attr.text.replace('\n','').replace('\xa0','')) for attr in now_table_header.findAll('th')]



now_new_table = pd.DataFrame(columns=now_attr_list)

row_marker = 0

for row in now_table_body.find_all('tr'):
    row_vals = [row_val.text.strip() for row_val in row.find_all('td')]
    for i, num in enumerate(row_vals):
        if ',' in num:
            row_vals[i] = int(num.replace(',', ''))
       
    
    new_row = pd.Series(row_vals, index = now_new_table.columns)
    now_new_table = now_new_table.append(new_row, ignore_index=True)


```

```python
now = datetime.now()
now_new_table.to_csv(f"./data/{now.day}-{now.month}-{now.year}.csv", sep='\t',encoding='utf-8', index = False)
```
